```python
import xarray as xr
import matplotlib.pyplot as plt
import cmocean as cm
```

```python
import openeo

connection = openeo.connect("openeo.dataspace.copernicus.eu")
```

See: https://documentation.dataspace.copernicus.eu/APIs/openEO/Collections.html

```python
connection.authenticate_oidc()
```

```python
# List collections available on the openEO back-end
connection.list_collection_ids()

# Get detailed metadata of a certain collection
#connection.describe_collection("SENTINEL2_L2A")
connection.describe_collection("SENTINEL3_SLSTR")

#connection.describe_collection("SENTINEL3_SYN_L2_SYN")
#connection.describe_collection("SENTINEL3_OLCI_L1B")
```

```python
datacube = connection.load_collection(
    "SENTINEL3_SLSTR",
    spatial_extent={"west": 120., "south": -15.5, "east": 125., "north": -12},
    temporal_extent = ["2023-08-01", "2023-09-01"],
    bands=["S7","S8","S9"],
    #max_cloud_cover=100,
)
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = xarray.load_dataset("../DATA/sentinel3_OLCI.nc")

# Convert xarray DataSet to a (bands, t, x, y) DataArray
data = ds[["B08", "B06", "B04"]].to_array(dim="bands")
```

```python
fig, (axrgb, axhist) = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))

# plot the data
data[{"t": 0}].plot.imshow(ax=axrgb)

# Plot the data histogram
data.plot.hist(bins=50, ax=axhist, histtype="stepfilled")

plt.show()

```
